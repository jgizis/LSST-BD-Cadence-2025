```python
import numpy as np
```

```python
#read the volume sum 
volume17=np.load('V17_volume_results.npy')
volume50=np.load('volumesum_V50.npy')
volume53=np.load('volumesum_V53.npy')
volume53_11yr=np.load('volumesum_V53_11yrs.npy')
```

```python
#fairly sloppy way of making type array
type=np.arange(20)
#leave as integers rather than make into L0, L1,... L9, T0, ..., T9
```

```python
#For Zenodo only, record as astropy table
#
# units are cubic parsecs, but I won't bother with that. Sorry purists! :)
```

```python
from astropy.table import Table
```

```python
t = Table([type,volume17, volume50, volume53, volume53_11yr],
           names=('Type','V1.7', 'V5.0', 'V5.3', 'V5.3_11yr'),
           meta={'name': 'Volume Sums'})
```
